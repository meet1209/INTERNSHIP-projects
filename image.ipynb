```python
img = r"/home/meet-kesur/Documents/15 days/download.jpeg"
img
```

```python
i_a = np.array(img)
i_a
```

```python
i_c = img.convert('L')
i_c
```

```python
from sklearn.decomposition import PCA
```

```python
p = PCA(n_components = 50)
p
```

```python
i_a1 = np.array(i_c)
p1 = p.fit_transform(i_a1)
p1
```

```python
import matplotlib.pyplot as plt
plt.imshow(p1)
plt.show()
```

```python
p2 = p.inverse_transform(p1)
p2
```

```python
plt.imshow(p2)
plt.show()
```
